```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
```

```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # optional: for ggplot-like style
print('OK to proceed')
```

```python
df_bnb=pd.read_csv('http://data.insideairbnb.com/singapore/sg/singapore/2020-03-21/visualisations/listings.csv')
df_bnb.tail()
```

```python
#let's check if the price dont have unrealistic value
df_bnb.sort_values(by='price', axis=0, ascending=False, inplace=True)
df_bnb.head(50)
```

```python
#as the value above 1000 sgd will crush the other prices, let's drop them
df_bnb.reset_index().inplace=True
```

```python
df_bnb.drop(df_bnb.index[:48], inplace=True)
```

```python
#let's format our dataset a bit
df_bnb.rename(columns={'neighbourhood':'Neighborhood'}, inplace=True)
df_bnb[['Neighborhood']].head()
```

```python
#Adding a column to better count the rows..
df_bnb['Count']=1
df_bnb['Count'].tail()
```

```python
df_bnb[['Neighborhood','Count']].head()
```

```python
#let's regroup the neighborhood to have a aggregated view
df_co=df_bnb.groupby('Neighborhood',axis=0).mean()
df_co.reset_index(inplace=True)
df_co
```

```python
#let's reformat the prices to weight them with the number of occurence (average price per neighborhood)
df_co['average_price']=(df_co['price']/df_co['Count']).round(2)
df_co['average_price'].head()
```

```python
#let's check if the avg prices make sense
df_co['average_price'].min()
```

```python
df_co['average_price'].max()
```

```python
df_co['average_price'].dtype
```

```python
df_co['average_price_neighborhood'] = df_co['Neighborhood']+': $' + df_co['average_price'].astype(str)
df_co.head()
```

```python
#we have a nice 1:10 relation between the min and max, we will be able to slice the map in  5 color intensity using a 2 scale
df_co
```

```python
!conda install -c conda-forge folium=0.5.0 --yes
import folium

print('Folium installed and imported!')
```

```python
#a quick google search gives us the location of Singapore
latitude = 1.3521
longitude = 103.8198
```

```python
!wget --quiet http://data.insideairbnb.com/singapore/sg/singapore/2020-03-21/visualisations/neighbourhoods.geojson -O map.geojson
    
print('GeoJSON file downloaded!')
singap = r'neighbourhoods.geojson' # geojson file
```

```python
#lets check whether we are righlty centered on Singapore
sg_map1 = folium.Map(location=[latitude, longitude], zoom_start=12)
sg_map1
```

```python
#let's apply the choropleth procedure 
threshold_scale = np.linspace(df_co['average_price'].min(),
                              df_co['average_price'].max(),
                              5, dtype=int)
threshold_scale = threshold_scale.tolist() # change the numpy array to a list
threshold_scale[-1] = threshold_scale[-1] + 1 # make sure that the last value of the list is greater than the maximum immigration

sg_map2 = folium.Map(location=[latitude, longitude], zoom_start=11, tiles='Mapbox Bright')
sg_map2.choropleth(
    geo_data=singap,
    data=df_co,
    columns=['Neighborhood', 'average_price'],
    key_on='feature.properties.name',
    threshold_scale=threshold_scale,
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Average price for Air BNB per district',
    reset=True
)
sg_map2
```

```python
sg_map3 = folium.Map(location=[latitude, longitude], zoom_start=12)
```

```python
#we can display the average price per neighborhood
sg_map = folium.Map(location=[latitude, longitude], zoom_start=12)

for lat, lng, label in zip(df_co.latitude, df_co.longitude, df_co.average_price_neighborhood):
    folium.features.CircleMarker(
        [lat, lng],
        radius=7,
        color='blue',
        fill=True,
        popup=label,
        fill_color='blue',
        fill_opacity=0.4
    ).add_to(sg_map3)

sg_map3
```
